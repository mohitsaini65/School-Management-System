```python
from faker import Faker
from datetime import datetime, timedelta
import random

fake = Faker("en_IN")
cursor = db.cursor()

# OPTIONAL: Clean previous attendance
cursor.execute("DELETE FROM attendance;")
cursor.execute("ALTER TABLE attendance AUTO_INCREMENT = 1;")
db.commit()

# Get student → class mapping
cursor.execute("SELECT student_id, class_id FROM enrollments;")
enrollments = cursor.fetchall()

# Define 3-month period
end_date = datetime.today()
start_date = end_date - timedelta(days=90)

current_date = start_date

total_records = 0

print("Generating attendance...")

while current_date <= end_date:
    for student_id, class_id in enrollments:

        status = random.choices(
            ["Present", "Absent"], 
            weights=[85, 15],  # 85% present
            k=1
        )[0]

        query = """
        INSERT INTO attendance (student_id, class_id, date, status)
        VALUES (%s, %s, %s, %s)
        """

        cursor.execute(query, (student_id, class_id, current_date.date(), status))
        total_records += 1

    db.commit()
    current_date += timedelta(days=1)

print(f"Inserted {total_records} attendance records successfully!")
```

```python
import mysql.connector
import random
from datetime import datetime, timedelta

db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Mohit@6565",
    database="school_mgmt"
)

cursor = db.cursor()

# Get all student IDs
cursor.execute("SELECT student_id FROM students")
students = [row[0] for row in cursor.fetchall()]

# Get all subject IDs
cursor.execute("SELECT subject_id FROM subjects")
subjects = [row[0] for row in cursor.fetchall()]

exam_types = ["Mid-Term", "Final"]
start_date = datetime(2024, 3, 1)

count = 0

for student in students:
    for subject in subjects:
        for exam in exam_types:
            
            marks = random.randint(30, 100)  # realistic range
            exam_date = start_date + timedelta(days=random.randint(0, 60))

            query = """
            INSERT INTO marks (student_id, subject_id, exam_type, marks_obtained, exam_date)
            VALUES (%s, %s, %s, %s, %s)
            """
            values = (student, subject, exam, marks, exam_date)
            cursor.execute(query, values)
            count += 1

db.commit()
print(f"Inserted {count} marks records successfully!")
```

```python
import mysql.connector
import random
from datetime import datetime, timedelta

db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Mohit@6565",
    database="school_mgmt"
)

cursor = db.cursor()

cursor.execute("SELECT student_id FROM students")
students = [row[0] for row in cursor.fetchall()]

count = 0

for student in students:

    amount = 30000  # annual fee
    status = random.choice(["Paid", "Pending"])

    # If paid, generate date between April 1 – June 30
    if status == "Paid":
        start = datetime(2024, 4, 1)
        paid_on = start + timedelta(days=random.randint(0, 90))
    else:
        paid_on = None  # not paid

    query = """
    INSERT INTO fees (student_id, amount, paid_on, status)
    VALUES (%s, %s, %s, %s)
    """

    cursor.execute(query, (student, amount, paid_on, status))
    count += 1

db.commit()
print(f"Inserted {count} fee records successfully!")
```
